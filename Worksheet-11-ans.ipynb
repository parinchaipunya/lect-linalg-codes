```python
import numpy as np
import matplotlib.pyplot as plt
```

## Problem 1

Consider the functions
$$
f(x) = \sin(2\pi x) \cos(\pi x)
$$
and
$$
g(x) = \frac{x + \sin(x)}{x + \cos(x)}
$$
over the interval $x \in [-\pi, \pi]$.

### Problem 1.1

Plot the two functions on the same axis.

```python
x = np.linspace(-np.pi,np.pi, 200)
f = np.sin(2 * np.pi * x) * np.cos(np.pi * x)
g = (x + np.sin(x)) / (x + np.cos(x))
```

```python
plt.plot(x,f, label="f(x)")
plt.plot(x,g, label="g(x)")
plt.legend()
plt.show()
```

### Problem 1.2

Plot the two functions side-by-side in two subplots.

```python
fig = plt.figure(figsize=(14,4))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.plot(x,f)
ax1.set_title("f(x)")
ax2.plot(x,g)
ax2.set_title("g(x)")
plt.show()
```

## Problem 2

Plot the surface of the function
$$
F(x,y) = \frac{\sin(x^{2} + 3y^{2})}{x^{2} + 3y^{2}}
$$
over $x,y \in [-2\pi,2\pi]$.

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-2*np.pi,2*np.pi,500)
y = np.linspace(-2*np.pi,2*np.pi,500)
X, Y = np.meshgrid(x,y)

W = X**2 + 3*Y**2
Z = np.sin(W) / W
```

```python
# Create a figure and an empty axis inside it.
fig = plt.figure(figsize=(10,3), constrained_layout=True)
ax = fig.add_subplot(111, projection='3d')

# Plot the surface inside the axis.
ax.plot_surface(X,Y,Z)


# Show the surface.
plt.show()
```
